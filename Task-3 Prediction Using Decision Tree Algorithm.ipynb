# Task-1 Prediction Using Supervised ML

### Importing Needed Packages

```python
import numpy as np
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

### Load the Dataset

```python
data = load_iris()
df = pd.DataFrame(data.data, columns = data.feature_names)
df['target'] = data.target
```

```python
df.head()
```

### Spliting data into Training and Test sets

```python
X_train, X_test, Y_train, Y_test = train_test_split(df[data.feature_names], df['target'], random_state = 0)
```

### Modeling the pattern

```python
#Make an instance of the model
clf = DecisionTreeClassifier(max_depth = 2, random_state = 0)
```

```python
#Train the model on the data
clf.fit(X_train, Y_train)
```

```python
#Predict the labels of unseen (test) data
clf.predict(X_test)
```

### Visualize Decision Tree

```python
tree.plot_tree(clf)
```

```python
fn = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
cn = ['setosa', 'versicolor', 'verginica']
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4 , 4), dpi = 300)

tree.plot_tree(clf, feature_names = fn, class_names = cn, filled = True)

fig.savefig('iris_tree.png')
```

```python
fig.savefig("decision_tree.png") # save decision tree 
```

# Type 2 Visualize A Decision Tree

### load the libraries

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
from sklearn import tree
import matplotlib.pyplot as plt
```

### Load the Data

```python
iris = datasets.load_iris()
```

```python
X = iris.data

y = iris.target
```

### Train Decision Tree

```python
#Create decision Tree classifier object
clf = DecisionTreeClassifier(random_state = 0)

#Train model

model = clf.fit(X , y)
```

```python
text_representation = tree.export_text(clf)
print(text_representation)
```

```python
with open("decision_tree.log", "w") as fout:
    fout.write(text_representation)
```

```python
fig = plt.figure(figsize = (25 , 20))
tree = tree.plot_tree(clf, feature_names = iris.feature_names, class_names = iris.target_names, filled = True)
```

```python
fig.savefig("decision_tree.png") # save decision tree 
```
